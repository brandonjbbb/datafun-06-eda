```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = sns.load_dataset('iris')
print(df.head(10))

```

```python
# Check the shape of the dataset (rows, columns)
print(df.shape)

# Display the data types of each column
print(df.dtypes)
```

```python
# Generate summary statistics for the dataset
print(df.describe())
# Plot histograms for all numerical columns
df.hist(figsize=(10, 8))
plt.tight_layout()
plt.show()
```

```python
# Plot histograms for all numerical columns
df.hist(figsize=(10, 8))
plt.tight_layout()
plt.show()
```

```python
# Plot a count plot for the 'species' column
sns.countplot(x='species', data=df)
plt.title('Distribution of Species')
plt.show()
```

```python
# Create a pairplot to visualize relationships between numerical features
sns.pairplot(df, hue='species')
plt.show()
```

```python
# Exclude non-numerical columns for correlation
corr_matrix = df.select_dtypes(include=['float64', 'int64']).corr()

# Create a heatmap to visualize the correlations
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Rename columns
df.rename(columns={'sepal_length': 'Sepal Length', 'sepal_width': 'Sepal Width'}, inplace=True)
```

```python
# Create the 'Petal Length/Width Ratio' column if it doesn't exist
df['Petal Length/Width Ratio'] = df['petal_length'] / df['petal_width']
print(df.head(5))

```

```python
# Create a boxplot to visualize the new feature by species
sns.boxplot(x='species', y='Petal Length/Width Ratio', data=df)
plt.title('Petal Length/Width Ratio by Species')
plt.show()
```
