```python
# Create Sepal Length/Width Ratio column
df['Sepal Length/Width Ratio'] = df['sepal_length'] / df['sepal_width']
```

```python
# Exclude non-numerical columns for correlation
corr_matrix = df.select_dtypes(include=['float64', 'int64']).corr()

# Create a heatmap to visualize the correlations
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

### Correlation Heatmap of Numerical Features

The heatmap below visualizes the correlation between the numerical features in the Iris dataset:

- **Sepal Length** and **Petal Length** have a positive correlation of 0.87, indicating that flowers with longer sepals also tend to have longer petals.
- **Petal Length** and **Petal Width** are highly correlated (0.96), suggesting that these two features are closely related and might be significant in distinguishing between species.
- The newly created **Sepal Length/Width Ratio** has a moderate positive correlation with **Sepal Length** (0.55) but shows a weak correlation with other features, indicating that it may offer unique insights into species classification.

This correlation matrix helps us understand the relationships between the features and identify the most strongly correlated pairs that could be important in further analysis or modeling.


```python
# Rename columns
df.rename(columns={'sepal_length': 'Sepal Length', 'sepal_width': 'Sepal Width'}, inplace=True)
```

```python
# Check if the 'Petal Length/Width Ratio' column exists in the DataFrame
print(df.columns)
```

```python
df['Petal Length/Width Ratio'] = df['petal_length'] / df['petal_width']
```

```python
# Create a boxplot to visualize the new feature by species
sns.boxplot(x='species', y='Petal Length/Width Ratio', data=df)
plt.title('Petal Length/Width Ratio by Species')
plt.show()
```

### Boxplot of Petal Length/Width Ratio by Species

The boxplot below shows the distribution of the **Petal Length/Width Ratio** across the three species:

- **Setosa** species have the lowest Petal Length/Width Ratio, with relatively little variation between the samples. This suggests that Setosa species have proportionally smaller petals compared to their width.
- **Versicolor** species display a wider range of values for the Petal Length/Width Ratio, indicating greater variability in petal dimensions within this species.
- **Virginica** species tend to have the highest Petal Length/Width Ratio, suggesting proportionally longer petals compared to the other species.

This boxplot highlights clear differences in petal proportions between species, particularly between Setosa and Virginica, which could be useful for distinguishing between them.


```python
sns.violinplot(x='species', y='sepal_length', data=df)

plt.title('Violin Plot of Sepal Length by Species')
plt.show()
```
